```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "mouse_drug_data.csv"
clinical_trial_data_to_load = "clinicaltrial_data.csv"

# Read the Mouse Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
clinical_df = pd.read_csv(clinical_trial_data_to_load)

# remove mouse g989 from both data sets
mouse_df = mouse_df[mouse_df['Mouse ID'] != 'g989']
clinical_df = clinical_df[clinical_df['Mouse ID'] != 'g989']

# Combine the data into a single dataset
mc_df = clinical_df.merge(mouse_df, left_on='Mouse ID', right_on='Mouse ID')

mc_df.head()
```

# Tumor Response to Treatment

## Mean

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint

# Grouped Data
grouped_dt = mc_df.groupby(by=['Drug','Timepoint'])

# Mean Tumor Volume
mean_tr = grouped_dt['Tumor Volume (mm3)'].mean()

# Convert to DataFrame
mean_tr_df = pd.DataFrame(mean_tr).reset_index()

# Preview DataFrame
mean_tr_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames (Mean)

mean_t_df_clean = mean_tr_df.pivot(index='Timepoint',columns='Drug')
mean_t_df_clean.head()
```

## Standard Deviation

```python
# Calculate the percent changes for each drug
# grab the 0 and 45 indices then calc

# (final tumor size - original tumor size) / original tumor size

# var = mean_t_df_clean.apply(lambda x: x.div(x.iloc[0]).mul(100))

# Display the data to confirm

# return first and last, apply pct change, drop index, reset * 100
first_last = mean_t_df_clean.iloc[[0,-1],].pct_change().drop(index=0).reset_index(drop=True) * 100

# transpose, reset index, drop column, rename, set index to Drug column
ptv_change = first_last.T.reset_index().drop(columns='level_0').rename(columns={0:'% Tumor Volume Change'}).set_index('Drug')
ptv_change
```

```python
# Store all Relevant Percent Changes into a Tuple
data = tuple(ptv_change['% Tumor Volume Change'])
ptv_change.plot(kind='bar',figsize=(20,15),fontsize=20,grid=True,width=0.8,alpha=.5)
plt.xlabel('Drug', size=30, labelpad=20)

plt.ylabel('% Tumor Volume Change', size=30, labelpad=20)

plt.title('Survival During Treatment',size=40, pad=20)
plt.legend(loc='best',prop={'size': 20})
plt.bar(ptv_change.index,ptv_change['% Tumor Volume Change'],color=['g','r','r','r','r','r','r','g','r','r'])

x = 0
for percent in data:
    plt.text(x, 0, str(round(percent,2))+'%', ha='center',color='black',size=22)
    x+=1        

plt.savefig("Graphs/Overall.png")
# Splice the data between passing and failing drugs


# Orient widths. Add labels, tick marks, etc. 


# Use functions to label the percentages of changes


# Call functions to implement the function calls


# Save the Figure


# Show the Figure
#fig.show()
```

# Trends and Analysis on Drug Efficacy

1. Of all the drugs tested, the data show that Capomulin and Ramicane show the most promise as possible treatments. On average, not only did these drugs halt further tumor growth, they reduced tumor volume by nearly 20 percent. 

2. Mice given Ketapril, Naftisol, and Stelasyn showed an increase in tumor size compared to the Placebo.  

3. In general, mice given Infubinol and Propriva died at a faster rate than mice given the Placebo. 

4. All drugs except for Ketapril showed a decrease in the spread of the cancer according to Metastatic Sites.
